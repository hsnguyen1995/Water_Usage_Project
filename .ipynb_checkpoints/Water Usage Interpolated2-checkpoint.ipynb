```python
import os
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from functools import reduce

file = "Resources/cleanest_data.csv"

water_df = pd.read_csv(file)
```

```python
# Display dataset from csv
water_df
```

```python
# Count how many individual counties there are in the dataset
counties = water_df["County"].drop_duplicates()
counties.count()
```

```python
# Potentially useful dataframe which sets county name and year as index
summary_by_county = pd.DataFrame(water_df.groupby(["County","Year"]).sum()[["Domestic total self-supplied withdrawals plus deliveries, in Mgal/d",
                                                                  "Industrial, Self-Supply Withdrawn (Mgal/day)", 
                                                                  "Industrial Total Consumption (Mgal/day)",
                                                                   "Livestock, Self-Supply, Freshwater (Mgal/day)",
                                                                   "Livestock Consumption Freshwater (Mgal/day)",
                                                                   "Irrigation, Crop, Self-Supply, Fresh (Mgal/day)",
                                                                   "Irrigation Crop Consumtion Freshwater (Mgal)"]])

# Find total water use by adding all subcategories of use together
summary_by_county = pd.DataFrame(summary_by_county["Domestic total self-supplied withdrawals plus deliveries, in Mgal/d"] +
                                 summary_by_county["Industrial, Self-Supply Withdrawn (Mgal/day)"] +
                                 summary_by_county["Industrial Total Consumption (Mgal/day)"] +
                                 summary_by_county["Livestock, Self-Supply, Freshwater (Mgal/day)"] +
                                 summary_by_county["Livestock Consumption Freshwater (Mgal/day)"] +
                                 summary_by_county["Irrigation, Crop, Self-Supply, Fresh (Mgal/day)"] +
                                 summary_by_county["Irrigation Crop Consumtion Freshwater (Mgal)"],
                                columns = ['Total Water Use (Mgal/day)'])

summary_by_county.head()
```

```python
# Find data for just Los Angeles County
LA_df = pd.DataFrame(water_df.loc[water_df["County"] == 'Los Angeles County',:])

LA_df.head()
```

```python
LA_df = LA_df.rename(columns={"Domestic total self-supplied withdrawals plus deliveries, in Mgal/d": 
                              "Total LA County Residential Use (Mgal/day)"})

LA_df["Total LA County Agricultural Use (Mgal/day)"] = LA_df["Livestock, Self-Supply, Freshwater (Mgal/day)"] + LA_df["Livestock Consumption Freshwater (Mgal/day)"] + LA_df["Irrigation, Crop, Self-Supply, Fresh (Mgal/day)"] + LA_df["Irrigation Crop Consumtion Freshwater (Mgal)"]
            
LA_df["Total LA County Industrial Use (Mgal/day)"] = LA_df["Industrial, Self-Supply Withdrawn (Mgal/day)"] + LA_df["Industrial Total Consumption (Mgal/day)"]
    

LA_df = LA_df[["Year", "Total LA County Residential Use (Mgal/day)", "Total LA County Industrial Use (Mgal/day)", "Total LA County Agricultural Use (Mgal/day)"]]

LA_df.set_index('Year')
```

```python
populations = pd.DataFrame(water_df.groupby("Year").sum()["Population Served in Thousands"])
populations = populations.rename(columns = {"Population Served in Thousands" : "Population Served in CA (in Thousands)"})

populations
```

```python
industrial_summary = pd.DataFrame(water_df.groupby("Year").sum()[["Industrial, Self-Supply Withdrawn (Mgal/day)", 
                                                                  "Industrial Total Consumption (Mgal/day)"]])

industrial_total = pd.DataFrame(industrial_summary["Industrial, Self-Supply Withdrawn (Mgal/day)"] +
                                industrial_summary["Industrial Total Consumption (Mgal/day)"],
                                columns = ['Total Industrial Use in CA (Mgal/day)'])

industrial_total
```

```python
agri_summary = pd.DataFrame(water_df.groupby("Year").sum()[["Livestock, Self-Supply, Freshwater (Mgal/day)",
                                                            "Livestock Consumption Freshwater (Mgal/day)",
                                                            "Irrigation, Crop, Self-Supply, Fresh (Mgal/day)",
                                                            "Irrigation Crop Consumtion Freshwater (Mgal)"]])

agri_total = pd.DataFrame(agri_summary["Livestock, Self-Supply, Freshwater (Mgal/day)"] + 
                                  agri_summary["Livestock Consumption Freshwater (Mgal/day)"]+
                                  agri_summary["Irrigation, Crop, Self-Supply, Fresh (Mgal/day)"] + 
                                  agri_summary["Irrigation Crop Consumtion Freshwater (Mgal)"], 
                                  columns = ['Total Agricultural Use in CA (Mgal/day)'])

agri_total
```

```python
residential_total = pd.DataFrame(water_df.groupby("Year").sum()
                                 ["Domestic total self-supplied withdrawals plus deliveries, in Mgal/d"])

residential_total = residential_total.rename(columns = {"Domestic total self-supplied withdrawals plus deliveries, in Mgal/d" :
                                                        "Total Residential Use in CA (Mgal/day)"})

residential_total
```

```python
data_frames = [populations, residential_total, industrial_total, agri_total]
total_state_data = reduce(lambda  left,right: pd.merge(left,right,on=['Year'],
                                            how='outer'), data_frames)

total_state_data = total_state_data.replace(0.00, np.nan)
total_state_data
```

```python
LA_df
```

```python
comparison_df = pd.merge(total_state_data, LA_df, on = "Year", how = "outer")
comparison_df.set_index('Year')

comparison_df
```

```python
# Make a new dataframe which "Year" is set as index
comparison2_df = comparison_df.set_index("Year")

# Set plot size
plt.figure(figsize =(10,7))

# Plot comparison2_df with the second column as the second y-axis
ax = comparison2_df.plot(secondary_y = ['Population Served in CA (in Thousands)'])

ax.set_ylabel('Water Use (Million Gallons per Day)')
ax.right_ax.set_ylabel('Population (Thousands)')

# Place legend outside of the plot
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.right_ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

ax.legend(loc='center left', bbox_to_anchor=(1.25, 0.5))
ax.right_ax.legend(loc='center left', bbox_to_anchor=(1.25, .8))

# Name title of the plot and x-axis
plt.title ("State vs. Los Angeles County Total Water Usage by Type")
plt.xlabel("Year")

plt.show()
```

```python
scatter_df = pd.DataFrame({"x" : total_state_data["Population Served in CA (in Thousands)"],
                           "y" : total_state_data["Total Residential Use in CA (Mgal/day)"],
                           "year" : total_state_data.index})

scatter_df2 = scatter_df.dropna()

# basic plot
p1=sns.regplot(data=scatter_df2, x="x", y="y", fit_reg=False, marker="o", color="skyblue")
 
# add annotations one by one with a loop
for line in range(0,scatter_df2.shape[0]):
     p1.text(scatter_df2.x[line]+0.2, scatter_df2.y[line], scatter_df2.group[line], horizontalalignment='left', size='medium', color='black', weight='semibold')

plt.show()

# plt.scatter(total_state_data["Population Served in CA (in Thousands)"], 
#             total_state_data["Total Residential Use in CA (Mgal/day)"])



# for year in total_state_data.index:
    
#     label = year
#     x = total_state_data["Population Served in CA (in Thousands)"]
#     y = total_state_data["Total Residential Use in CA (Mgal/day)"]
    
#     plt.annotate(label, (x,y))
    
# plt.show()
```
